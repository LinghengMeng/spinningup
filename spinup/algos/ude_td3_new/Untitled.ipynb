```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_randomNet_log_data(data_df, suptitle=''):
    marker = '.'
    markersize = 0.8#1#0.8

    fig, axes = plt.subplots(1,3,sharey=True)
    fig.set_figwidth(18)
    fig.set_figheight(4)
    fig.suptitle(suptitle, fontsize=18)
    axes[0].plot(data_df['LBDPredError'], marker, markersize=markersize)
    axes[0].set_xlabel('Steps')
    axes[0].set_ylabel('Prediction Error')
    axes[0].set_title('Lazy Bernoulli Dropout Prediction Error')

    axes[1].plot(data_df['BDPredError'], marker, markersize=markersize)
    axes[1].set_xlabel('Steps')
    axes[1].set_title('Bernoulli Dropout Prediction Error')

    axes[2].plot(data_df['BEPredError'], marker, markersize=markersize)
    axes[2].set_xlabel('Steps')
    axes[2].set_title('Bootstrapped Ensemble Prediction Error')
    plt.subplots_adjust(top=0.85)
    plt.savefig('./figures/{}_step_log_prediction_error.jpg'.format(suptitle), dpi=300)

    fig, axes = plt.subplots(1,3,sharey=True)
    fig.set_figwidth(18)
    fig.set_figheight(4)
    fig.suptitle(suptitle, fontsize=18)
    axes[0].plot(data_df['LBDUnc'], marker, markersize=markersize)
    axes[0].set_xlabel('Steps')
    axes[0].set_ylabel('Estimated Uncertainty')
    axes[0].set_title('Lazy Bernoulli Dropout Estimated Uncertainty')

    axes[1].plot(data_df['BDUnc'], marker, markersize=markersize)
    axes[1].set_xlabel('Steps')
    axes[1].set_title('Bernoulli Dropout Estimated Uncertainty')

    axes[2].plot(data_df['BEUnc'], marker, markersize=markersize)
    axes[2].set_xlabel('Steps')
    axes[2].set_title('Bootstrapped Ensemble Estimated Uncertainty')
    plt.subplots_adjust(top=0.85)
    plt.savefig('./figures/{}_step_log_estimated_uncertainty.jpg'.format(suptitle), dpi=300)


    fig, axes = plt.subplots(1,3,sharey=True)
    fig.set_figwidth(18)
    fig.set_figheight(4)
    fig.suptitle(suptitle, fontsize=18)
    axes[0].plot(data_df['LBDLoss'], marker, markersize=markersize)
    axes[0].set_xlabel('Steps')
    axes[0].set_ylabel('Training Loss')
    axes[0].set_title('Lazy Bernoulli Dropout Training Loss')

    axes[1].plot(data_df['BDLoss'], marker, markersize=markersize)
    axes[1].set_xlabel('Steps')
    axes[1].set_title('Bernoulli Dropout Training Loss')

    axes[2].plot(data_df['BELoss'], marker, markersize=markersize)
    axes[2].set_xlabel('Steps')
    axes[2].set_title('Bootstrapped Ensemble Dropout Training Loss')
    plt.subplots_adjust(top=0.85)
    plt.savefig('./figures/{}_step_log_training_loss.jpg'.format(suptitle), dpi=300)
```

```python
def plot_epsodic_log_data_rft(data_df, suptitle='', fig_name=''):
    fig, axes = plt.subplots(1,4)
    fig.set_figwidth(24)
    fig.set_figheight(4)
    fig.suptitle(suptitle, fontsize=18)

    axes[0].plot(data_df['AverageEpLBDEstiUnc'])
    axes[0].plot(data_df['AverageEpBDEstiUnc'])
    axes[0].plot(data_df['AverageEpBEEstiUnc'])
    axes[0].set_title('Uncertainty')
    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Estimated Uncertainty')
    axes[0].legend(['LBD', 'BD', 'BE'])

    axes[1].plot(data_df['AverageEpLBDPredError'])
    axes[1].plot(data_df['AverageEpBDPredError'])
    axes[1].plot(data_df['AverageEpBEPredError'])
    axes[1].set_title('Prediction Error')
    axes[1].set_xlabel('Epochs')
    axes[1].set_ylabel('Prediction Error')
    axes[1].legend(['LBD', 'BD', 'BE'])

    axes[2].plot(data_df['AverageEpLBDTrainLoss'])
    axes[2].plot(data_df['AverageEpBDTrainLoss'])
    axes[2].plot(data_df['AverageEpBETrainLoss'])
    axes[2].set_title('Training Loss')
    axes[2].set_xlabel('Epochs')
    axes[2].set_ylabel('Training Loss')
    axes[2].legend(['LBD', 'BD', 'BE'])
    
    axes[3].plot(data_df['AverageTestEpRet'])
    axes[3].plot(data_df['AverageEpRet'])
    axes[3].set_title('Accumulated Reward')
    axes[3].set_xlabel('Epochs')
    axes[3].set_ylabel('Accumulated Reward')
    axes[3].legend(['Test', 'Train'])
    
    plt.subplots_adjust(top=0.85)
    plt.savefig('./figures/{}'.format(fig_name), dpi=300)
#     plt.tight_layout(True)
```

```python
def plot_epsodic_log_data(data_df):
    fig, axes = plt.subplots(1,3)
    fig.set_figwidth(18)

    axes[0].plot(data_df['AverageTestEpRet'])
    axes[0].plot(data_df['AverageEpRet'])
    axes[0].set_title('Accumulated Reward')
    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Accumulated Reward')
    axes[0].legend(['Test', 'Train'])

    axes[1].plot(data_df['AverageQ1Vals'])
    axes[1].plot(data_df['AverageQ2Vals'])
    axes[1].set_title('Estimated Q-value')
    axes[1].set_xlabel('Epochs')
    axes[1].set_ylabel('Estimated Q-value')
    axes[1].legend(['Q1', 'Q2'])

    axes[2].plot(data_df['LossPi'])
    axes[2].plot(data_df['LossPi'])
    axes[2].plot(data_df['LossPi'])
    axes[2].set_title('Pi Loss')
    axes[2].set_xlabel('Epochs')
    axes[2].set_ylabel('Training Loss')
    axes[2].legend(['LBD', 'BD', 'BE'])

    plt.tight_layout(True)
```
