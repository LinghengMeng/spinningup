```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = r"C:\Users\pc-admin\Github Repos\spinningup\data\2019-04-16_ude_td3_HalfCheetah_v2_2L_300H_200Ep_3S_Drop_Unc_Delayed_Dropout_0_5_Con_Delayed_DMasks\2019-04-16_16-53-14-ude_td3_HalfCheetah_v2_2L_300H_200Ep_3S_Drop_Unc_Delayed_Dropout_0_5_Con_Delayed_DMasks_s3\dropout_uncertainty.txt"
data_df = pd.read_csv(data_path,  sep='\t')

for i in range(10):
#     fig, ax = plt.subplots()
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4) ,sharey=True)
    ax1.plot(np.arange(data_df.shape[0]), data_df['Obs{}_var'.format(i)])
    ax1.set_title('Obs{}_var'.format(i))
    
    ax2.plot(np.arange(data_df.shape[0]), data_df['Obs{}_cov'.format(i)])
    ax2.set_title('Obs{}_cov'.format(i))
    
    plt.tight_layout()
```
