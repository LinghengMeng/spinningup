```python
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def pick_task_progress_log(task_name, progress_files_dir_list):
    logs = []
    for f_dir in progress_files_dir_list:
        if task_name in f_dir:
            logs.append(f_dir)
    return logs
```

```python
def read_data(RoboschoolAnt_v1_files):
    dfs = []
    for f_dir in RoboschoolAnt_v1_files:
#         print(f_dir)
        dfs.append(pd.read_csv(f_dir,  sep='\t'))
    return dfs

def plot_return(RoboschoolAnt_v1_dfs, task_name='', legend=[], fig_name=''):
    fig, axes = plt.subplots(1,2, sharey=True)
    fig.set_figwidth(14)
    fig.suptitle('{}'.format(task_name), fontsize=18)
    for i in range(len(RoboschoolAnt_v1_dfs)):
        axes[0].plot(RoboschoolAnt_v1_dfs[i]['AverageEpRet'])
        axes[1].plot(RoboschoolAnt_v1_dfs[i]['AverageTestEpRet'])
    axes[0].set_ylabel('Accumulated Reward')
    axes[0].set_xlabel('Epochs')
    axes[1].set_xlabel('Epochs')
    axes[0].legend(legend)
    axes[1].legend(legend)
    axes[0].set_title('Train Episod Return')
    axes[1].set_title('Test Episod Return')
    plt.subplots_adjust(top=0.85)
    fig.savefig('./figures/{}'.format(fig_name), dpi=300)
```

## Both Pi and Q Using Dropout

```python
data_dir = r"C:\Users\Lingheng\Google Drive\Compute_Canada\ddpg_dropout_both_policy_and_q_using_dropout"
sub_data_dirs = os.listdir(data_dir)
```

```python
progress_files_dir = []
for sub_dir in sub_data_dirs:
    print(sub_dir)
    for d in os.listdir(os.path.join(data_dir, sub_dir)):
        # print(d)
        progress_files_dir.append(os.path.join(data_dir, sub_dir, d, 'progress.txt'))
        # t = glob.glob()
```

```python
tasks = ['RoboschoolAnt-v1', 'RoboschoolHalfCheetah-v1', 'RoboschoolHopper-v1','RoboschoolWalker2d-v1']
RoboschoolAnt_v1_files = pick_task_progress_log('RoboschoolAnt-v1', progress_files_dir)
RoboschoolHalfCheetah_v1_files = pick_task_progress_log('RoboschoolHalfCheetah-v1', progress_files_dir)
RoboschoolHopper_v1_files = pick_task_progress_log('RoboschoolHopper-v1', progress_files_dir)
RoboschoolWalker2d_v1_files = pick_task_progress_log('RoboschoolWalker2d-v1', progress_files_dir)
```

```python

RoboschoolAnt_v1_dfs = read_data(RoboschoolAnt_v1_files)
plot_return(RoboschoolAnt_v1_dfs, task_name='RoboschoolAnt_v1', legend=['0', '0.01', '0.05', '0.1', '0.2'], 
            fig_name='Both_Pi_and_Q_Using_Dropout_RoboschoolAnt_v1.jpg')

RoboschoolHalfCheetah_v1_dfs = read_data(RoboschoolHalfCheetah_v1_files)
plot_return(RoboschoolHalfCheetah_v1_dfs, task_name='RoboschoolHalfCheetah-v1', legend=['0', '0.01', '0.05', '0.1', '0.2'], 
            fig_name='Both_Pi_and_Q_Using_Dropout_RoboschoolHalfCheetah_v1.jpg')

RoboschoolHopper_v1_dfs = read_data(RoboschoolHopper_v1_files)
plot_return(RoboschoolHopper_v1_dfs, task_name='RoboschoolHopper-v1', legend=['0', '0.01', '0.05', '0.1', '0.2'], 
            fig_name='Both_Pi_and_Q_Using_Dropout_RoboschoolHopper_v1.jpg')

RoboschoolWalker2d_v1_dfs = read_data(RoboschoolWalker2d_v1_files)
plot_return(RoboschoolWalker2d_v1_dfs, task_name='RoboschoolWalker2d-v1', legend=['0', '0.01', '0.05', '0.1', '0.2'], 
            fig_name='Both_Pi_and_Q_Using_Dropout_RoboschoolWalker2d_v1.jpg')
```
