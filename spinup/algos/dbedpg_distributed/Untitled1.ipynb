```python
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_data(RoboschoolAnt_v1_0_1_0_75_df, RoboschoolAnt_v1_0_1_1_df, 
              RoboschoolAnt_v1_0_20_0_75_df, RoboschoolAnt_v1_0_20_1_df, task='RoboschoolAnt-v1'):
    fig, axes = plt.subplots(1,4, sharey=True)
    fig.set_figwidth(18)
    fig.suptitle('{}'.format(task), fontsize=18)
    axes[0].plot(RoboschoolAnt_v1_0_1_0_75_df['AverageEpRet'])
    axes[0].plot(RoboschoolAnt_v1_0_1_1_df['AverageEpRet'])
    axes[0].plot(RoboschoolAnt_v1_0_20_0_75_df['AverageEpRet'])
    axes[0].plot(RoboschoolAnt_v1_0_20_1_df['AverageEpRet'])
    axes[0].legend(['1 actor, 0.75 replay buffer', '1 actor, 1 replay buffer', 
                    '20 actor, 0.75 replay buffer', '20 actor, 1 replay buffer'])
    axes[0].set_title('AverageEpRet')

    axes[1].plot(RoboschoolAnt_v1_0_1_0_75_df['AverageTestEpRetAC0'])
    axes[1].plot(RoboschoolAnt_v1_0_1_1_df['AverageTestEpRetAC0'])
    axes[1].legend(['1 actor, 0.75 replay buffer', '1 actor, 1 replay buffer'])
    axes[1].set_title('Single Actor: AverageTestEpRet')

    for i in range(20):
        axes[2].plot(RoboschoolAnt_v1_0_20_0_75_df['AverageTestEpRetAC{}'.format(i)])
    axes[2].set_title('Ensemble: AverageTestEpRet\n 0.75 replay buffer')

    for i in range(20):
        axes[3].plot(RoboschoolAnt_v1_0_20_1_df['AverageTestEpRetAC{}'.format(i)])
    axes[3].set_title('Ensemble: AverageTestEpRet\n 1 replay buffer')
    plt.subplots_adjust(top=0.8)
    plt.savefig('./figures/{}', dpi=300)
```
