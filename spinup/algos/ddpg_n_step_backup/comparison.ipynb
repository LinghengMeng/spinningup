```python
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_data(data_frames=[Hopper_v2_1step_df, Hopper_v2_5step_df], 
              legend=['1-step', '5-step'], task_name='', fig_name=''):
    fig, axes = plt.subplots(1,2,sharey=True)
    fig.set_figwidth(14)
    fig.suptitle('{}'.format(task_name), fontsize=18)
    for data in data_frames:
        axes[0].plot(data['AverageEpRet'])
    axes[0].set_title('AverageEpRet')
    axes[0].legend(legend)
    
    for data in data_frames:
        axes[1].plot(data['AverageTestEpRet'])
    axes[1].set_title('AverageTestEpRet')
    axes[1].legend(legend)
    fig.savefig('./figures/{}'.format(fig_name), dpi=300)
```

```python
def plot_data_TD3(task_name='Ant-v2'):
    fig, axes = plt.subplots(1,2,sharey=True)
    fig.set_figwidth(14)
    fig.suptitle('{}'.format(task_name), fontsize=18)
    axes[0].plot(Ant_v2_1step_df['AverageEpRet'])
    axes[0].plot(Ant_v2_3step_df['AverageEpRet'])
    axes[0].plot(Ant_v2_5step_df['AverageEpRet'])
    axes[0].set_title('AverageEpRet')
    axes[0].legend(['1-step', '3-step', '5-step'])

    axes[1].plot(Ant_v2_1step_df['AverageTestEpRet'])
    axes[1].plot(Ant_v2_3step_df['AverageTestEpRet'])
    axes[1].plot(Ant_v2_5step_df['AverageTestEpRet'])
    axes[1].set_title('AverageTestEpRet')
    axes[1].legend(['1-step', '3-step', '5-step'])
```

```python
plot_data_TD3(task_name='Ant-v2')
```
