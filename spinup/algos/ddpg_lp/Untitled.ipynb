```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_learning_progress_episodic(data_df):
    fig, axes = plt.subplots(1,5)
    fig.set_figwidth(18)

    axes[0].plot(data_df['AverageEpActorVar'])
    axes[0].plot(data_df['AverageEpActorVarFirst'])
    axes[0].plot(data_df['AverageEpActorVarSecond'])
    axes[0].set_title('Episodic Variance on Actor')
    axes[0].legend(['All', 'First Half', 'Second Half'])

    axes[1].plot(data_df['AverageEpCriticVar'])
    axes[1].plot(data_df['AverageEpCriticVarFirst'])
    axes[1].plot(data_df['AverageEpCriticVarSecond'])
    axes[1].set_title('Episodic Variance on Critic')
    axes[1].legend(['All', 'First Half', 'Second Half'])

    axes[2].plot(data_df['AverageEpActorLPNorm'])
    axes[2].set_title('Episodic Learning Progress Norm\n on Actor')

    axes[3].plot(data_df['AverageEpCriticLPNorm'])
    axes[3].set_title('Episodic Learning Progress Norm\n on Critic')
    
    axes[4].plot(data_df['AverageEpRet'])
    axes[4].plot(data_df['AverageTestEpRet'])
    axes[4].set_title('Episodic Return')
    axes[4].legend(['Train', 'Test'])
```

```python
def plot_learning_progress_step(data_df):
    marker = '.'
    markersize = 0.01#1#0.8

    fig, axes = plt.subplots(1,5)
    fig.set_figwidth(18)

    axes[0].plot(data_df['VarAll'], marker, markersize=markersize)
    axes[0].set_title('Variance on Whole Window')

    axes[1].plot(data_df['VarFirstHalf'], marker, markersize=markersize)
    axes[1].set_title('Variance on 1st Half Window')

    axes[2].plot(data_df['VarSecondHalf'], marker, markersize=markersize)
    axes[2].set_title('Variance on 2nd Half Window')

    axes[3].plot(data_df['VarChange'], marker, markersize=markersize)
    axes[3].set_title('Variance Change')

    axes[4].plot(data_df['L2LP'], marker, markersize=markersize)
    axes[4].set_title('L2 Norm on Learning Progress')
```
