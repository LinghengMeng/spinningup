```python
def plot_data(data_frames=[], 
              legend=['1-step', '5-step'], task_name='', fig_name=''):
    fig, axes = plt.subplots(1,2,sharey=True)
    fig.set_figwidth(14)
    fig.suptitle('{}'.format(task_name), fontsize=18)
    for data in data_frames:
        axes[0].plot(data['AverageEpRet'])
    axes[0].set_title('AverageEpRet')
    axes[0].legend(legend)
    
    for data in data_frames:
        axes[1].plot(data['AverageTestEpRet'])
    axes[1].set_title('AverageTestEpRet')
    axes[1].legend(legend)
    fig.savefig('./figures/{}'.format(fig_name), dpi=300)
```

```python
def plot_data_comparison(df_group_1=[], df_group_2=[],
              legend=['1-step', '5-step'], task_name='', fig_name=''):
    fig, axes = plt.subplots(1,2,sharey=True)
    fig.set_figwidth(14)
    fig.suptitle('{}'.format(task_name), fontsize=18)
    
    epochs=199
    # Group 1
    ep_ret_1 = np.zeros((len(df_group_1), epochs))
    test_ep_ret_1 = np.zeros((len(df_group_1), epochs))
    for df_i, df_tmp in enumerate(df_group_1):
        ep_ret_1[df_i, :len(df_tmp['AverageEpRet'])] = df_tmp['AverageEpRet']
        test_ep_ret_1[df_i, :len(df_tmp['AverageTestEpRet'])]= df_tmp['AverageTestEpRet']
    avg_ep_ret_1 = np.mean(ep_ret_1, axis=0)
    std_ep_ret_1 = np.std(ep_ret_1, axis=0)
    avg_test_ep_ret_1 = np.mean(test_ep_ret_1, axis=0)
    std_test_ep_ret_1 = np.std(test_ep_ret_1, axis=0)
    
    # Group 2
    ep_ret_2 = np.zeros((len(df_group_2), epochs))
    test_ep_ret_2 = np.zeros((len(df_group_2), epochs))
    for df_i, df_tmp in enumerate(df_group_2):
        ep_ret_2[df_i, :len(df_tmp['AverageEpRet'])] = df_tmp['AverageEpRet']
        test_ep_ret_2[df_i, :len(df_tmp['AverageTestEpRet'])]= df_tmp['AverageTestEpRet']
    avg_ep_ret_2 = np.mean(ep_ret_2, axis=0)
    std_ep_ret_2 = np.std(ep_ret_2, axis=0)
    avg_test_ep_ret_2 = np.mean(test_ep_ret_2, axis=0)
    std_test_ep_ret_2 = np.std(test_ep_ret_2, axis=0)
    
    alpha = 0.25
    
    axes[0].plot(avg_ep_ret_1)
    axes[0].fill_between(np.arange(epochs), avg_ep_ret_1-std_ep_ret_1, avg_ep_ret_1+std_ep_ret_1, alpha=alpha)
    axes[0].plot(avg_ep_ret_2)
    axes[0].fill_between(np.arange(epochs), avg_ep_ret_2-std_ep_ret_2, avg_ep_ret_2+std_ep_ret_2, alpha=alpha)
    axes[0].set_title('AverageEpRet')
    axes[0].legend(legend)
    axes[0].grid(True, linestyle=':')

    axes[1].plot(avg_test_ep_ret_1)
    axes[1].fill_between(np.arange(epochs), avg_test_ep_ret_1-std_test_ep_ret_1, avg_test_ep_ret_1+std_test_ep_ret_1, alpha=alpha)
    axes[1].plot(avg_test_ep_ret_2)
    axes[1].fill_between(np.arange(epochs), avg_test_ep_ret_2-std_test_ep_ret_2, avg_test_ep_ret_2+std_test_ep_ret_2, alpha=alpha)
    axes[1].set_title('AverageTestEpRet')
    axes[1].legend(legend)
    axes[1].grid(True, linestyle=':')

    fig.savefig('./figures/{}'.format(fig_name), dpi=300)
```
