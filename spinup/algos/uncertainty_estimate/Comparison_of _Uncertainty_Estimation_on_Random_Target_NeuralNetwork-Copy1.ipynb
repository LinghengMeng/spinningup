## Walker2d-v2

```python
marker = '.'
markersize = 0.5
plt.figure()
plt.plot(data_df['r'], marker, markersize=markersize)
```

```python
marker = '.'
markersize = 0.1

# Loss
f, axes = plt.subplots(1, 3)
f.set_figwidth(18)
f.set_figheight(4)

axes[0].plot(data_df['MLPLoss'], marker, markersize=markersize)
axes[0].set_title('Lazy Bernoulli Dropout (MLP) Loss')
axes[0].tick_params(axis='x', labelrotation=30)
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Loss')

axes[1].plot(data_df['BernDropLoss'], marker, markersize=markersize)
axes[1].set_title('Bernoulli Dropout Loss')
axes[1].tick_params(axis='x', labelrotation=30)
axes[1].set_xlabel('Steps')
axes[1].set_ylabel('Loss')

axes[2].plot(data_df['BootsEnsembleAvgLoss'], marker, markersize=markersize)
axes[2].set_title('Bootstrapped Ensemble Loss')
axes[2].tick_params(axis='x', labelrotation=30)
axes[2].set_xlabel('Steps')
axes[2].set_ylabel('Loss')
f.savefig('./realistic_example_loss.jpg', dpi=300)

# Uncertainty
f, axes = plt.subplots(1, 3, sharey=True)
f.set_figwidth(18)
f.set_figheight(4)

axes[0].plot(data_df['LazyBernDropUnc'], marker, markersize=markersize)
axes[0].set_title('Lazy Bernoulli Dropout Uncertainty')
axes[0].tick_params(axis='x', labelrotation=30)
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Uncertainty')

axes[1].plot(data_df['BernDropUnc'], marker, markersize=markersize)
axes[1].set_title('Bernoulli Dropout Uncertainty')
axes[1].tick_params(axis='x', labelrotation=30)
axes[1].set_xlabel('Steps')

axes[2].plot(data_df['BootstrappedEnsembleUnc'], marker, markersize=markersize)
axes[2].set_title('Bootstrapped Ensemble Uncertainty')
axes[2].tick_params(axis='x', labelrotation=30)
axes[2].set_xlabel('Steps')
f.savefig('./realistic_example_uncertainty.jpg', dpi=300)

# Uncertainty Histogram
f, axes = plt.subplots(1, 3, sharey=True, sharex=True)
f.set_figwidth(18)
f.set_figheight(3)
num_bins=50
axes[0].hist(data_df['LazyBernDropUnc'], num_bins, facecolor='blue', alpha=0.5)
axes[0].set_xlabel('Uncertainty')
axes[0].set_ylabel('Count')
axes[0].set_title('Lazy Bernoulli Dropout Uncertainty Histogram')

axes[1].hist(data_df['BernDropUnc'], num_bins, facecolor='blue', alpha=0.5)
axes[1].set_xlabel('Uncertainty')
axes[1].set_title('Bernoulli Dropout Uncertaity Histogram')

axes[2].hist(data_df['BootstrappedEnsembleUnc'], num_bins, facecolor='blue', alpha=0.5)
axes[2].set_xlabel('Uncertainty')
axes[2].set_title('Bootstrapped Ensemble Uncertainty Histogram')
f.savefig('./realistic_example_uncertainty_hist.jpg', dpi=300)
```

```python
stages = int(data_df['LazyBernDropUnc'].shape[0] / 2000)
stages
```

```python

stages = 20

f, axes = plt.subplots(stages, 3, sharey=True, sharex=True)
f.set_figwidth(18)
f.set_figheight(stages*3)
for i in range(stages):
    num_bins=50
    axes[i][0].hist(data_df['LazyBernDropUnc'][i*2000:(i+1)*2000], num_bins, facecolor='blue', alpha=0.5)
    axes[i][0].set_title('LazyBernDropUnc: [step_{} , step_{}]'.format(i*2000,(i+1)*2000))

    axes[i][1].hist(data_df['BernDropUnc'][i*2000:(i+1)*2000], num_bins, facecolor='blue', alpha=0.5)
    axes[i][1].set_title('BernDropUnc: [step_{} , step_{}]'.format(i*2000,(i+1)*2000))

    axes[i][2].hist(data_df['BootstrappedEnsembleUnc'][i*2000:(i+1)*2000], num_bins, facecolor='blue', alpha=0.5)
    axes[i][2].set_title('BootstrappedEnsembleUnc: [step_{} , step_{}]'.format(i*2000,(i+1)*2000))

f.savefig('./realistic_example_uncertainty_hist_change.jpg', dpi=300)
```
