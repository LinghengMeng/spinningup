```python
def plot_log_data(data_df, fig_name=''):
    marker = '.'
    markersize = 1#0.8

    data_set_size = data_df['LazyBernDropPredError'].shape[0]#10000
    training_steps = data_df['LazyBernDropPredError'].shape[0]

    # Prediction error
    f, axes = plt.subplots(1, 3, sharey=True)
    f.set_figwidth(18)
    f.set_figheight(4)

    axes[0].plot(data_df['LazyBernDropPredError'], marker, markersize=markersize)
    axes[0].set_title('Lazy Bernoulli Dropout Prediction Error')
    axes[0].tick_params(axis='x', labelrotation=30)
    axes[0].set_xlabel('Steps')
    axes[0].set_ylabel('Prediction Error')
    # axes[0].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[0].text(data_set_size/2, 0.8, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center',wrap=True)
    # axes[0].text((training_steps-data_set_size)/2+data_set_size, 0.8, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)

    axes[1].plot(data_df['BernDropPredError'], marker, markersize=markersize)
    axes[1].set_title('Bernoulli Dropout Prediction Error')
    axes[1].tick_params(axis='x', labelrotation=30)
    axes[1].set_xlabel('Steps')
    # axes[1].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[1].text(data_set_size/2, 0.8, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[1].text((training_steps-data_set_size)/2+data_set_size, 0.8, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)

    axes[2].plot(data_df['BootstrappedEnsemblePredError'], marker, markersize=markersize)
    axes[2].set_title('Bootstrapped Ensemble Prediction Error')
    axes[2].tick_params(axis='x', labelrotation=30)
    axes[2].set_xlabel('Steps')
    # axes[2].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[2].text(data_set_size/2, 0.8, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[2].text((training_steps-data_set_size)/2+data_set_size, 0.8, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)
    f.savefig('./realistic_example_collecting_at_beginning_pred_error_{}.jpg'.format(fig_name), dpi=300)

    # Loss
    f, axes = plt.subplots(1, 3, sharey=True)
    f.set_figwidth(18)
    f.set_figheight(4)

    axes[0].plot(data_df['MLPLoss'], marker, markersize=markersize)
    axes[0].set_title('Lazy Bernoulli Dropout (MLP) Loss')
    axes[0].tick_params(axis='x', labelrotation=30)
    axes[0].set_xlabel('Steps')
    axes[0].set_ylabel('Loss')
    # axes[0].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[0].text(data_set_size/2, 0.2, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[0].text((training_steps-data_set_size)/2+data_set_size, 0.2, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)

    axes[1].plot(data_df['BernDropLoss'], marker, markersize=markersize)
    axes[1].set_title('Bernoulli Dropout Loss')
    axes[1].tick_params(axis='x', labelrotation=30)
    axes[1].set_xlabel('Steps')
    # axes[1].set_ylabel('Loss')
    # axes[1].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[1].text(data_set_size/2, 25, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[1].text((training_steps-data_set_size)/2+data_set_size, 25, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)

    axes[2].plot(data_df['BootsEnsembleAvgLoss'], marker, markersize=markersize)
    axes[2].set_title('Bootstrapped Ensemble Loss')
    axes[2].tick_params(axis='x', labelrotation=30)
    axes[2].set_xlabel('Steps')
    # axes[2].set_ylabel('Loss')
    # axes[2].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[2].text(data_set_size/2, 0.2, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[2].text((training_steps-data_set_size)/2+data_set_size, 0.2, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)
    f.savefig('./realistic_example_collecting_at_beginning_loss_{}.jpg'.format(fig_name), dpi=300)

    # Uncertainty
    f, axes = plt.subplots(1, 3, sharey=True)
    f.set_figwidth(18)
    f.set_figheight(4)

    axes[0].plot(data_df['LazyBernDropUnc'], marker, markersize=markersize)
    axes[0].set_title('Lazy Bernoulli Dropout Uncertainty')
    axes[0].tick_params(axis='x', labelrotation=30)
    axes[0].set_xlabel('Steps')
    axes[0].set_ylabel('Uncertainty')
    # axes[0].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[0].text(data_set_size/2, 0.1, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[0].text((training_steps-data_set_size)/2+data_set_size, 0.1, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)

    axes[1].plot(data_df['BernDropUnc'], marker, markersize=markersize)
    axes[1].set_title('Bernoulli Dropout Uncertainty')
    axes[1].tick_params(axis='x', labelrotation=30)
    axes[1].set_xlabel('Steps')
    # axes[1].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[1].text(data_set_size/2, 0.1, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[1].text((training_steps-data_set_size)/2+data_set_size, 0.1, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)

    axes[2].plot(data_df['BootstrappedEnsembleUnc'], marker, markersize=markersize)
    axes[2].set_title('Bootstrapped Ensemble Uncertainty')
    axes[2].tick_params(axis='x', labelrotation=30)
    axes[2].set_xlabel('Steps')
    # axes[2].axvline(x=data_set_size, linestyle='--', linewidth=2, color='r')
    # axes[2].text(data_set_size/2, 0.1, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
    # axes[2].text((training_steps-data_set_size)/2+data_set_size, 0.1, 'Only Training', fontsize=12, ha='center', va='center', wrap=True)
    f.savefig('./realistic_example_collecting_at_beginning_uncertainty_{}.jpg'.format(fig_name), dpi=300)

    # Uncertainty Histogram
    f, axes = plt.subplots(1, 3, sharey=True, sharex=True)
    f.set_figwidth(18)
    f.set_figheight(3)
    num_bins=50
    axes[0].hist(data_df['LazyBernDropUnc'], num_bins, facecolor='blue', alpha=0.5)
    axes[0].set_xlabel('Uncertainty')
    axes[0].set_ylabel('Count')
    axes[0].set_title('Lazy Bernoulli Dropout Uncertainty Histogram')

    axes[1].hist(data_df['BernDropUnc'], num_bins, facecolor='blue', alpha=0.5)
    axes[1].set_xlabel('Uncertainty')
    axes[1].set_title('Bernoulli Dropout Uncertaity Histogram')

    axes[2].hist(data_df['BootstrappedEnsembleUnc'], num_bins, facecolor='blue', alpha=0.5)
    axes[2].set_xlabel('Uncertainty')
    axes[2].set_title('Bootstrapped Ensemble Uncertainty Histogram')
    f.savefig('./realistic_example_collecting_at_beginning_uncertainty_hist_{}.jpg'.format(fig_name), dpi=300)
```

```python
marker = '.'
markersize = 0.5

data_set_size = data_df['LazyBernDropPredError'].shape[0]

# Prediction error
f, axes = plt.subplots(1, 3, sharey=True)
f.set_figwidth(18)
f.set_figheight(4)

axes[0].plot(data_df['LazyBernDropPredError'], marker, markersize=markersize)
axes[0].set_title('Lazy Bernoulli Dropout Prediction Error')
axes[0].tick_params(axis='x', labelrotation=30)
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Prediction Error')
axes[0].text(data_set_size/2, 1.7, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center',wrap=True)

axes[1].plot(data_df['BernDropPredError'], marker, markersize=markersize)
axes[1].set_title('Bernoulli Dropout Prediction Error')
axes[1].tick_params(axis='x', labelrotation=30)
axes[1].set_xlabel('Steps')
axes[1].text(data_set_size/2, 1.7, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)

axes[2].plot(data_df['BootstrappedEnsemblePredError'], marker, markersize=markersize)
axes[2].set_title('Bootstrapped Ensemble Prediction Error')
axes[2].tick_params(axis='x', labelrotation=30)
axes[2].set_xlabel('Steps')
axes[2].text(data_set_size/2, 1.7, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)

f.savefig('./realistic_example_training_and_collecting_pred_error.jpg', dpi=300)

# Loss
f, axes = plt.subplots(1, 3)
f.set_figwidth(18)
f.set_figheight(4)

axes[0].plot(data_df['MLPLoss'], marker, markersize=markersize)
axes[0].set_title('Lazy Bernoulli Dropout (MLP) Loss')
axes[0].tick_params(axis='x', labelrotation=30)
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Loss')
axes[0].text(data_set_size/2, 0.3, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)

axes[1].plot(data_df['BernDropLoss'], marker, markersize=markersize)
axes[1].set_title('Bernoulli Dropout Loss')
axes[1].tick_params(axis='x', labelrotation=30)
axes[1].set_xlabel('Steps')
axes[1].set_ylabel('Loss')
axes[1].text(data_set_size/2, 100, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)

axes[2].plot(data_df['BootsEnsembleAvgLoss'], marker, markersize=markersize)
axes[2].set_title('Bootstrapped Ensemble Loss')
axes[2].tick_params(axis='x', labelrotation=30)
axes[2].set_xlabel('Steps')
axes[2].set_ylabel('Loss')
axes[2].text(data_set_size/2, 0.25, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
f.savefig('./realistic_example_training_and_collecting_loss.jpg', dpi=300)

# Uncertainty
f, axes = plt.subplots(1, 3, sharey=True)
f.set_figwidth(18)
f.set_figheight(4)

axes[0].plot(data_df['LazyBernDropUnc'], marker, markersize=markersize)
axes[0].set_title('Lazy Bernoulli Dropout Uncertainty')
axes[0].tick_params(axis='x', labelrotation=30)
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Uncertainty')
axes[0].text(data_set_size/2, 0.8, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)

axes[1].plot(data_df['BernDropUnc'], marker, markersize=markersize)
axes[1].set_title('Bernoulli Dropout Uncertainty')
axes[1].tick_params(axis='x', labelrotation=30)
axes[1].set_xlabel('Steps')
axes[1].text(data_set_size/2, 0.8, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)

axes[2].plot(data_df['BootstrappedEnsembleUnc'], marker, markersize=markersize)
axes[2].set_title('Bootstrapped Ensemble Uncertainty')
axes[2].tick_params(axis='x', labelrotation=30)
axes[2].set_xlabel('Steps')
axes[2].text(data_set_size/2, 0.8, 'Data Collecting\n& Training', fontsize=12, ha='center', va='center', wrap=True)
f.savefig('./realistic_example_training_and_collecting_uncertainty.jpg', dpi=300)

# Uncertainty Histogram
f, axes = plt.subplots(1, 3, sharey=True, sharex=True)
f.set_figwidth(18)
f.set_figheight(3)
num_bins=50
axes[0].hist(data_df['LazyBernDropUnc'], num_bins, facecolor='blue', alpha=0.5)
axes[0].set_xlabel('Uncertainty')
axes[0].set_ylabel('Count')
axes[0].set_title('Lazy Bernoulli Dropout Uncertainty Histogram')

axes[1].hist(data_df['BernDropUnc'], num_bins, facecolor='blue', alpha=0.5)
axes[1].set_xlabel('Uncertainty')
axes[1].set_title('Bernoulli Dropout Uncertaity Histogram')

axes[2].hist(data_df['BootstrappedEnsembleUnc'], num_bins, facecolor='blue', alpha=0.5)
axes[2].set_xlabel('Uncertainty')
axes[2].set_title('Bootstrapped Ensemble Uncertainty Histogram')
f.savefig('./realistic_example_training_and_collecting_uncertainty_hist.jpg', dpi=300)
```
