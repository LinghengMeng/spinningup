# Investigate Uncertainty Change

**Obs0_var** is the summation of diagnal of covariance matrix.

**Obs0_cov** is the summation of up triangle of covariance matrix.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### HalfCheetah-v2
#### UD3: no_delay_train + 10000 start_steps

```python
data_path = "../../data/2019-03-31_ud3_HalfCheetah_3S_2L_200Ep_1Drop_GSP_1Con_0Low_no_delay_train/2019-03-31_19-40-02-ud3_HalfCheetah_3S_2L_200Ep_1Drop_GSP_1Con_0Low_no_delay_train_s3/uncertainty.txt"
data_df = pd.read_csv(data_path,  sep='\t')

for i in range(10):
#     fig, ax = plt.subplots()
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4) ,sharey=True)
    ax1.plot(np.arange(data_df.shape[0]), data_df['Obs{}_var'.format(i)])
    ax1.set_title('Obs{}_var'.format(i))
    
    ax2.plot(np.arange(data_df.shape[0]), data_df['Obs{}_cov'.format(i)])
    ax2.set_title('Obs{}_cov'.format(i))
    
    plt.tight_layout()
```
